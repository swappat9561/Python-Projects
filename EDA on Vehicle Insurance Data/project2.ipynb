ii. Outliers
Generate a summary of count of all the outliers column wise
(Hint1: for outlier treatment use IQR method as follows:
For example: for a column X calculate Q1 = 25th percentile and Q3 = 75th percentile 
then IQR = Q3 – Q1 ) then to check outlier, anything lower than a Q1 – 1.5IQR or 
greater than Q3 + 1.5 IQR would be an outlier


Hint2: For getting percentile value, explore pd.describe() function)

```python
# summary of count of all the outliers

df1.describe()
```

```python
df2.describe()
```

```python
# checking outliers for age

print ('Outliers in age are :')

Q1=df1.describe().loc["25%","Age"]

Q3=df1.describe().loc["75%","Age"]

IQR=Q3-Q1

print(df1.loc[df1["Age"]<(Q1-IQR*1.5),"Age"].count())

print(df1.loc[df1["Age"]>(Q3+IQR*1.5),"Age"].count())
```

```python
# checking outliers for annual premium (in Rs)

print ('\nOutliers in Annual Premium(Rs) : ')

Q1=df2.describe().loc["25%","Annual Premium(Rs)"]

Q3=df2.describe().loc["75%","Annual Premium(Rs)"]

IQR=Q3-Q1

print(df2.loc[df2["Annual Premium(Rs)"]<(Q1-IQR*1.5),"Annual Premium(Rs)"].count())

print(df2.loc[df2["Annual Premium(Rs)"]>(Q3+IQR*1.5),"Annual Premium(Rs)"].count())
```

```python
print('Replacing the outlier value for numeric columns of annual premium by mean: ')

df2.loc[df2["Annual Premium(Rs)"]>(Q3+IQR*1.5),"Annual Premium(Rs)"]=df2["Annual Premium(Rs)"].mean()

print(df2.loc[df2["Annual Premium(Rs)"]>(Q3+IQR*1.5),"Annual Premium(Rs)"].count())

print ('Now there are no outliers in annual premium (in Rs)')
```

```python
# checking for outliers in vintage

print ('\nOutliers in vintage are :')

Q1=df2.describe().loc["25%","Vintage"]

Q3=df2.describe().loc["75%","Vintage"]

IQR=Q3-Q1

print(df2.loc[df2["Vintage"]<(Q1-IQR*1.5),"Vintage"].count())

print(df2.loc[df2["Vintage"]>(Q3+IQR*1.5),"Vintage"].count()) 

print('There are no outliers in vintage column')
```
