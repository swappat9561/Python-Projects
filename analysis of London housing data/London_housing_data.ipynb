# convert data type of date column to Date-time format 

```python
df.dtypes
```

```python
# date has object data type so convert it to date time 

df.date=pd.to_datetime(df.date)
```

```python
df.dtypes   # data type changed successfully
```

# Add new column year to dataframe which contains year only

```python
df['year']=df.date.dt.year    # dt.year means from datetime year kadha fakta
```

```python
df
```

```python
df.insert(2,'month',df.date.dt.month)  

# syntax to insert month column
# df.insert(index,column name,contains of new column)
```

```python
df
```

# remove column year and month from data frame

```python
df.drop(['year','month'],axis=1,inplace=True)   
```
